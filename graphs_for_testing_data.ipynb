```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

# LAKE PP02 100,000
# Read in File as pandas dataframe
file_data = pd.read_csv("lake/lake_pp02_100000.csv")
lake_pp02_100000_df = pd.DataFrame(columns= file_data.iloc[0,:], data=file_data.iloc[1:,].as_matrix())

#PLOTTING  
fig, ax = plt.subplots(1,1, figsize=(10,6))
x_h = pd.to_numeric(lake_pp02_100000_df.index.values)
pp02_hy = pd.to_numeric(lake_pp02_100000_df.l.values)
ax.set_title("Lake PP02 100,000")
ax.set_xlabel("Iteration Number")
ax.set_ylabel("Iteration Score")
ax.plot(x_h , pp02_hy);
```
